```python
# Problem 1: Bank Account Create a class representing a bank account with attributes like account number, account holder name, and balance. Implement methods to deposit and withdraw money from the account.
class bank_account:
    def account_details(self,acc_number,name,balance):
        self.account_number=acc_number
        self.account_holder_name=name
        self.account_balance=balance
    def deposit(self,money):
        if money>0:
            self.account_balance+=money
            print(f"deposit money is {money} and total balance is {self.account_balance}")
        else:
            print("please enter valid amount to deposit in your account")
    def withdraw(self,money):
        if self.account_balance>=money:
            self.account_balance-=money
            print(f"withdraw amount is {money} and left balance is {self.account_balance}")
        else:
            print("withdraw is not possible")
abc=bank_account()
abc.account_details(6483268,'mitali',5000)
abc.deposit(1000)
abc.withdraw(8000)
```

```python
# Problem 3: Vehicle Rental Create a class representing a vehicle rental system. Implement methods to rent a vehicle, return a vehicle, and display available vehicles.
class VehicleRentalSystem:
    def __init__(self):
        self.available_vehicles = {}
        self.rented_vehicles = {}

    def add_vehicle(self, vehicle_id, vehicle_type):
        if vehicle_id not in self.available_vehicles:
            self.available_vehicles[vehicle_id] = vehicle_type
            print(f"Added {vehicle_type} with ID {vehicle_id} to available vehicles.")
        else:
            print(f"Vehicle with ID {vehicle_id} already exists.")

    def rent_vehicle(self, vehicle_id, customer_name):
        if vehicle_id in self.available_vehicles:
            vehicle_type = self.available_vehicles.pop(vehicle_id)
            self.rented_vehicles[vehicle_id] = customer_name
            print(f"Rented {vehicle_type} with ID {vehicle_id} to {customer_name}.")
        else:
            print(f"Vehicle with ID {vehicle_id} is not available for rent.")

    def return_vehicle(self, vehicle_id):
        if vehicle_id in self.rented_vehicles:
            customer_name = self.rented_vehicles.pop(vehicle_id)
            vehicle_type = self.available_vehicles[vehicle_id] if vehicle_id in self.available_vehicles else "Unknown"
            print(f"Returned {vehicle_type} with ID {vehicle_id} from {customer_name}.")
        else:
            print(f"Vehicle with ID {vehicle_id} is not rented out.")

    def display_available_vehicles(self):
        print("Available Vehicles:")
        for vehicle_id, vehicle_type in self.available_vehicles.items():
            print(f"ID: {vehicle_id}, Type: {vehicle_type}")

rental_system = VehicleRentalSystem()

rental_system.add_vehicle("V001", "Sedan")
rental_system.add_vehicle("V002", "SUV")

rental_system.rent_vehicle("V001", "John Doe")

rental_system.display_available_vehicles()

rental_system.return_vehicle("V001")

rental_system.display_available_vehicles()
```

```python
# Problem 4: Library Catalog Create classes representing a library and a book. Implement methods to add books to the library, borrow books, and display available books.

class Library:
    def __init__(self):
        self._books = []

    def add_book(self, new_book):
        self._books.append(new_book)

    def display_books(self):
        if self._books:  # if list of books is not empty
            print("The books we have made available in our library are:\n")
            for book in self._books:
                print(book)
        else:
            print("Sorry, we have no books available in the library at the moment")

    def borrow_book(self, requested_book):
        if requested_book in self._books:
            print("You have now borrowed \"%s\" " % requested_book)
            self._books.remove(requested_book)

        else:
            print("Sorry, \"%s\" is not there in our library at the moment" % requested_book)
abc=Library()
abc.add_book('pushp')
abc.display_books()
abc.borrow_book('sumer')
```

```python
# Problem 5: Product Inventory Create classes representing a product and an inventory system. Implement methods to add products to the inventory, update product quantity, and display available products.
class product:
    def __init__(self):
        self.inventory = {}
    def add_item(self, item_id, item_name, stock_count, price):
        self.inventory[item_id] = {"item_name": item_name, "stock_count": stock_count, "price": price}

    def update_item(self, item_id, stock_count, price):
        if item_id in self.inventory:
            self.inventory[item_id]["stock_count"] = stock_count
            self.inventory[item_id]["price"] = price
        else:
            print("Item not found in inventory.")

    def display_available_products(self, item_id):
        if item_id in self.inventory:
            item = self.inventory[item_id]
            return f"Product Name: {item['item_name']}, Stock Count: {item['stock_count']}, Price: {item['price']}"
        else:
            return "Item not found in inventory."
abc=product()
abc.add_item("c1","battery",80,1000)
abc.add_item("c2","keyboard",100,500)
abc.add_item("c3","mouse",200,900)
abc.add_item("c4","monitor",90,5000)
abc.update_item("c2",100,900)
print(abc.display_available_products("c3"))
print(abc.display_available_products("c1"))
print(abc.display_available_products("c2"))
print(abc.display_available_products("c4"))
```

```python
# Problem 6: Shape Calculation Create a class representing a shape with attributes like length, width, and height. Implement methods to calculate the area and perimeter of the shape.
class ShapeCalculation:
    def __init__(self, length, width, height, shape_type):
        self.length = length
        self.width = width
        self.height = height
        self.shape_type = shape_type
        self.area_value = 0
        self.perimeter_value = 0

    def calculate_area(self):
        if self.shape_type == "rectangle":
            self.area_value = self.length * self.width
        elif self.shape_type == "triangle":
            self.area_value = 0.5 * (self.width * self.height)
        elif self.shape_type == "square":
            self.area_value = self.length ** 2
        else:
            print("Area is not found for the given shape.")
        return self.area_value

    def calculate_perimeter(self):
        if self.shape_type == "rectangle":
            self.perimeter_value = 2 * (self.length + self.width)
        elif self.shape_type == "triangle":
            self.perimeter_value = self.length + self.width + self.height
        elif self.shape_type == "square":
            self.perimeter_value = 4 * self.length
        else:
            print("Perimeter is not found for the given shape.")
        return self.perimeter_value

abc = ShapeCalculation(3, 4, 5, "triangle")
area = abc.calculate_area()
perimeter = abc.calculate_perimeter()
print(f"Area: {area}")
print(f"Perimeter: {perimeter}")
```

```python
# Problem 8: Email Management Create a class representing an email with attributes like sender, recipient, and subject. Implement methods to send an email and display email details.
class GmailOps:
    def __init__(self):
        self.emails = []  # Use a list to store emails

    def send_email(self, recipient, subject, body, sender):
        email = {
            'recipient': recipient,
            'subject': subject,
            'body': body,
            'sender': sender
        }
        self.emails.append(email)

    def display_emails(self):
        for email in self.emails:
            print("Recipient:", email['recipient'])
            print("Subject:", email['subject'])
            print("Body:", email['body'])
            print("Sender:", email['sender'])
            print("-------------")


gmail = GmailOps()

gmail.send_email('abc', 'Late for school', 'Hi, I am late due to some reason', 'pqr')

gmail.display_emails()
```

```python
# Problem 10: ToDo List Create a class representing a ToDo list with attributes like tasks and due dates. Implement methods to add tasks, mark tasks as completed, and display pending tasks.
class ToDo:
    def __init__(self):
        self.tasks = dict()
    
    def add_task(self, todo_item, due_date):
        if todo_item not in self.tasks:
            self.tasks[todo_item] = due_date
    
    def mark_task_as_completed(self, todo_item):
        if todo_item in self.tasks:
            # You can mark a task as completed in various ways.
            # For this example, I'll simply remove the task from the dictionary.
            del self.tasks[todo_item]
    
    def display_pending_tasks(self):
        print("Pending Tasks:")
        for todo_item, due_date in self.tasks.items():
            print(f"{todo_item} (Due on: {due_date})")

my_todo = ToDo()
my_todo.add_task("Buy groceries", "2023-10-15")
my_todo.add_task("Do laundry", "2023-10-17")

my_todo.display_pending_tasks()  

my_todo.mark_task_as_completed("Buy groceries") 

my_todo.display_pending_tasks() 
```
